## Imports

```python
## TODO
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

## Load data

```python
## TODO
# load data
df = pd.read_excel('worldcities.xlsx')
df.head()
```

 ## Add column

:<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این مرحله با توجه به ستون جمعیت یک ستون جدید تحت عنوان Population Level   اضافه کنید به طوری که به 4 دسته زیر  بر اساس وضعیت هر داده اختصاص داده می شودند.
</br>
    <div dir="ltr">
Low: 0-25%
    </br>
Mid: 25-50%

    </br>

```python
## TODO
percentiles = [0, 0.25, 0.50, 0.70, 1]
labels = ['Low', 'Mid', 'High', 'Over']
df['Population Level'] = pd.qcut(df['population'], q=percentiles, labels=labels)

df
```

## Preproccess

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این بخش باید ستون های خالی داده ها راحذف کرده و در نهایت تمامی ستون ها را به داده های عددی تبدیل کنید به جز ستون اضافه شده
</font>
</p>
</div>

```python
# TODO
df = df.dropna()

df
```

```python
df.dtypes
```

```python
df_without_population_level = df.loc[:, df.columns != 'Population Level']

df_object_columns = df_without_population_level.select_dtypes(['object']).columns
df_without_population_level[df_object_columns] = df_without_population_level.select_dtypes(['object']).apply(lambda x: LabelEncoder().fit_transform(x))
df_without_population_level['Population Level'] = df['Population Level']
numeric_df = df_without_population_level
numeric_df
```

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 حال باید داده های به دست آمده از قسمت قبلی را نرمال سازی کنید.
</font>
</p>
</div>

```python

cols_to_normalize = numeric_df.columns.difference(['Population Level'])
numeric_df[cols_to_normalize] = MinMaxScaler().fit_transform(numeric_df[cols_to_normalize])
normalized_df = numeric_df
normalized_df
```

## Split

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
داده ها را به دو بخش تست و آموزش با نسبت 80 به 20 تقسیم کنید.
</font>
</p>
</div>

```python
train_df, test_df = train_test_split(normalized_df, test_size=0.20, random_state=42)
```

## Regression

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای این قسمت باید ستون Population  را به عنوان برچسب درنظر گرفته و از بقیه ستون هابه عنوان فیچر استفاده کنید
</font
</p>
</div>

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
باید از رگرسیون خطی
،درجه 1 و درجه 2 استفاده کرده و در هر مورد دقت و خطای میانگین مربعات را بدست آورده و نتایج را باهم مقایسه کنید.</font
</p>
    
</div>

```python
x_train = train_df.drop(['population', 'Population Level'], axis=1)
y_train = train_df['population']
x_test = test_df.drop(['population', 'Population Level'], axis=1)
y_test = test_df['population']
```

```python
# Linear Regression (1-degree)
linear_regression_instance = LinearRegression()
linear_regression_instance.fit(x_train, y_train)

y_pred1_train = linear_regression_instance.predict(x_train)
y_pred1_test = linear_regression_instance.predict(x_test)

mse1_train = mean_squared_error(y_train, y_pred1_train)
mse1_test = mean_squared_error(y_test, y_pred1_test)


print("1-degree linear regression\n")
print("train MSE error:", mse1_train)
print("test MSE error:", mse1_test)
print("train accuracy:", r2_score(y_train, y_pred1_train))
print("test accuracy:", r2_score(y_test, y_pred1_test))
```

```python
# Linear Regression (2-degree)
linear_regression_instance2 = LinearRegression()
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

linear_regression_instance2.fit(x_train_poly, y_train)
y_pred2_train = linear_regression_instance2.predict(x_train_poly)
y_pred2_test = linear_regression_instance2.predict(x_test_poly)

mse2_train = mean_squared_error(y_train, y_pred2_train)
mse2_test = mean_squared_error(y_test, y_pred2_test)


print("1-degree linear regression\n")
print("train MSE error:", mse2_train)
print("test MSE error:", mse2_test)
print("train accuracy:", r2_score(y_train, y_pred2_train))
print("test accuracy:", r2_score(y_test, y_pred2_test))
```

### Decision Tree (Entropy)

```python
x_train = train_df.drop(['population', 'Population Level'], axis=1)
y_train = train_df['Population Level']
x_test = test_df.drop(['population', 'Population Level'], axis=1)
y_test = test_df['Population Level']
```

```python
## TODO
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier(criterion='entropy', max_depth=12)

dt.fit(x_train, y_train)

y_pred_train_dt = dt.predict(x_train)
y_pred_test_dt = dt.predict(x_test)


# Calculate accuracy
accuracy_train = accuracy_score(y_train, y_pred_train_dt)
accuracy_test = accuracy_score(y_test, y_pred_test_dt)
print(f"Decision Tree Evaluation:\naccuracy on train set: {accuracy_train:.2f}\naccuracy on test set: {accuracy_test:.2f}")
```

### Random Forest (Entropy)

```python
# TODO
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, max_leaf_nodes=400, n_jobs=-1,criterion='entropy')
rf.fit(x_train,y_train)

y_pred_train = rf.predict(x_train)
accuracy_train = accuracy_score(y_train, y_pred_train)
y_pred_test =rf.predict(x_test)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Random Forest Evaluation:\naccuracy on train set: {accuracy_train:.2f}\naccuracy on test set: {accuracy_test:.2f}")
```

### KNN K=2
